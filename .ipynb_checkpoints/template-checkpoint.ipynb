# Classical ML Model

```python
import pandas as pd
import numpy as np
import jieba
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    ConfusionMatrixDisplay
)
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
```

```python
def chinese_tokenizer(text):
    return jieba.lcut(str(text))
```

```python
vectorizer = TfidfVectorizer(
    tokenizer=chinese_tokenizer,
    max_features=5000
)
```

### Train and Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    data["text"],
    data["polarity"],
    test_size=0.2,
    random_state=42,
    stratify=data["polarity"]
)
```

### Model Training

```python
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
clf = LogisticRegression(max_iter=1000, class_weight='balanced', solver='lbfgs')
clf.fit(X_train_tfidf, y_train)
```

### Model Evaluation

```python
y_pred = clf.predict(X_test_tfidf)
print("=== Classification Report ===")
print(classification_report(y_test, y_pred, digits=4))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot(cmap='Blues', xticks_rotation=45)
plt.title("Confusion Matrix")
plt.show()

feature_names = vectorizer.get_feature_names_out()
coefs = clf.coef_
```

#### Save Model and Vectorizer

```python
joblib.dump(clf, "./models/lr_model/lr_model.pkl")
joblib.dump(vectorizer, "./models/lr_model/tfidf_vectorizer.pkl")
```

#### Load Model

```python
lr_model = joblib.load("./models/lr_model/lr_model.pkl")
vectorizer = joblib.load("./models/lr_model/tfidf_vectorizer.pkl")
```

#### Inference Test

```python
new_text = ["产品做工精细，使用体验很好"]
X_new = vectorizer.transform(new_text)
proba = lr_model.predict_proba(X_new)[0]
pred_idx = np.argmax(proba)
pred_label = lr_model.classes_[pred_idx]

print(f"Probability {dict(zip(lr_model.classes_, proba.round(3)))}")
```

# Transformer-Based Model (BERT)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import TrainingArguments, Trainer
import evaluate
```

### Data Preparation

```python
label2id = {'negative': 0, 'neutral': 1, 'positive': 2}
id2label = {v: k for k, v in label2id.items()}

data['label'] = data['polarity'].map(label2id)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data['text'],
    data['label'],
    test_size=0.2,
    random_state=42,
    stratify=data['label']
)
```

```python
train_df = pd.DataFrame({"text": X_train, "label": y_train})
test_df = pd.DataFrame({"text": X_test, "label": y_test})
```

```python
train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(test_df)
```

### Load Pre-Trained Model and Tokenizer

```python
model_name = "bert-base-multilingual-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(data['polarity'].unique()))
```

```python
def preprocess_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

tokenized_train = train_dataset.map(preprocess_function, batched=True)
tokenized_test = test_dataset.map(preprocess_function, batched=True)
```

### Tokeniz data

```python
def preprocess_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

tokenized_train = train_dataset.map(preprocess_function, batched=True)
tokenized_test = test_dataset.map(preprocess_function, batched=True)
```

### Trainer

```python
accuracy = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = logits.argmax(axis=-1)
    return accuracy.compute(predictions=predictions, references=labels)

training_args = TrainingArguments(
    output_dir="./bert_output",
    do_train=True,
    do_eval=True,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    learning_rate=2e-5,
    logging_dir="./logs",
    logging_steps=10
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    compute_metrics=compute_metrics
)
```

### Model Training

```python
trainer.train()
```

```python
results = trainer.evaluate()
print(results)
```

```python
model.config.id2label = id2label
model.config.label2id = label2id
```

### Save Model

```python
path = "./models/bert_sentiment_model"

model.save_pretrained(path)
tokenizer.save_pretrained(path)
```

### Load Model

```python
tokenizer = AutoTokenizer.from_pretrained(path)
model = AutoModelForSequenceClassification.from_pretrained(path)
```

### Inference Test

```python
from transformers import pipeline

sentiment_pipeline = pipeline(
    "text-classification",
    model=model,
    tokenizer=tokenizer,
    return_all_scores=True
)

text = "这个手机的电池续航简直是个笑话"
preds = sentiment_pipeline(text)

print(preds)
```
